```python
df_final_copy = df_final1.copy()
```

```python
def local(x):
    name = ""
    if "West" in x or "East" in x.lower():
        a = x.split(" ")
        for s in range(0,len(a)-1):
            name = name+"s"
    else:
        return x
```

```python
df_final_copy["locality"] = df_final_copy["locality"].apply(lambda x:local(x))
df_final_copy.head()
```
